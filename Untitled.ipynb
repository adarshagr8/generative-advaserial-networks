<h1>Generative Advaserial Networks</h1><br>
</h4>This network trains a generator to generate MNIST images to fool the discriminator network.</h4>

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras.models import Model
from keras.layers import Input, Dense,Conv2D,BatchNormalization,Conv2DTranspose,Reshape,Dropout,Flatten
from keras import backend as K
import keras
from IPython.display import Image
from keras.utils import plot_model
```

```python
df=pd.read_csv('train.csv')
train=df.drop(columns=['label']).to_numpy().reshape((-1,28,28,1))/255
label=df['label'].to_numpy()
```

```python
VEC_SIZE=100
GEN_DROPOUT=0.25
GEN_KER_SIZE=(3,3)
GEN_STRIDE=(2,2)
GEN_LOSS='mse'
GEN_OPT=keras.optimizers.RMSprop(lr=5e-3, rho=0.9, epsilon=None, decay=0.0)


DISC_DROPOUT=0.35
DISC_STRIDE=(2,2)
DISC_KER_SIZE=(3,3)
DISC_LOSS='mse'
DISC_OPT=keras.optimizers.RMSprop(lr=2.5e-3, rho=0.9, epsilon=None, decay=0.0)

```

```python
def generator():
    a=Input((VEC_SIZE,))
    x=Dense((7*7*128))(a)
    x=Reshape((7,7,128))(x)
    x=Dropout(GEN_DROPOUT)(x)
    x=Conv2DTranspose(32,kernel_size=GEN_KER_SIZE,strides=GEN_STRIDE,padding='same',activation='relu')(x)
    x=BatchNormalization()(x)
    x=Dropout(GEN_DROPOUT)(x)
    x=Conv2DTranspose(1,kernel_size=GEN_KER_SIZE,strides=GEN_STRIDE,padding='same',activation='sigmoid')(x)
    model=Model(a,x)
    model.compile(loss=GEN_LOSS,optimizer=GEN_OPT,metrics=['accuracy'])
    return model
    
```

```python
def discriminator():
    a=Input((28,28,1))
    x=Conv2D(32,DISC_KER_SIZE,strides=DISC_STRIDE,padding='same',activation='relu')(a)
    x=BatchNormalization()(x)
    x=Dropout(DISC_DROPOUT)(x)
    x=Conv2D(64,DISC_KER_SIZE,strides=DISC_STRIDE,padding='same',activation='relu')(x)
    x=BatchNormalization()(x)
    x=Dropout(DISC_DROPOUT)(x)
    x=Flatten()(x)
    x=Dense((1024))(x)
    x=Dropout(GEN_DROPOUT)(x)
    x=Dense((1))(x)
    model=Model(a,x)
    model.compile(loss=DISC_LOSS,optimizer=DISC_OPT,metrics=['accuracy'])
    return model
```

```python
def advaserial(generator,discriminator):
    discriminator.trainable=False
    a=Input((VEC_SIZE,))
    model=Model(a,discriminator(generator(a)))
    model.compile(loss=DISC_LOSS,optimizer=DISC_OPT,metrics=['accuracy'])
    return model
```

```python
def train_discriminator(discriminator,epochs=1,batch_size=20000):
    for i in range(epochs):
        loc=np.random.randint(0,42000,batch_size//2)
        images=train[loc]
        label0=np.ones(images.shape[0])
        noise=np.random.random((batch_size//2,28,28,1))
        label1=np.ones(noise.shape[0])
        x=np.concatenate((images,noise), axis=0)
        y=np.concatenate((label0,label1), axis=0)
        discriminator.train_on_batch(x,y)
```

```python
def train_advaserial(advaserial,epochs=1,batch_size=20000):
    for i in range(epochs):
        x=np.random.random((batch_size,VEC_SIZE))
        y=np.ones((batch_size))
        advaserial.train_on_batch(x,y)
        
```

```python
GEN=generator()
DISC=discriminator()
ADV=advaserial(GEN,DISC)
```

```python
plot_model(GEN, to_file='GEN.png',show_shapes=True)
plot_model(DISC, to_file='DISC.png',show_shapes=True)
plot_model(ADV, to_file='ADV.png',show_shapes=True)
```
